# upload geoip data (geopi_block_ipv4 and geoip_country_location) to shared 

```scala
/** Always import Tetration App Lib **/
import com.tetration.apps.IO
import com.tetration.apps.DataTaps
import collection.JavaConverters._
import java.util.Calendar
//import sqlContext.implicits._

var files = IO.list(sqlContext, "/shared/geoip_block_ipv4/")
val ipblock = IO.read(sqlContext, "/shared/geoip_block_ipv4/", "CSV")
ipblock.registerTempTable("geoip_block")
ipblock.printSchema();
files = IO.list(sqlContext, "/shared/geoip_country_loc/")
val iploc = IO.read(sqlContext, "/shared/geoip_country_loc/", "CSV")
iploc.registerTempTable("geoip_loc")
iploc.printSchema();
```

# load 2 CSV files and register dataframe as tempoary table
# and just run sql statement for matching IP addresses and get Country name 

```scala
sqlContext.sql("""SELECT a.network, b.continent_name, b.country_name 
FROM geoip_block as a, geoip_loc as b 
WHERE a.geoname_id = b.geoname_id and a.network like '64.104%'""").show(30)
```
